```python
def write_to_csv(tweets,filename):
    headers = ['id', 'full_text', 'created_at','lang', 'sent_class']

    with open(filename + '.csv' , 'w' , encoding="utf-8", newline='') as csvfile:
        writer = csv.writer(csvfile,delimiter=',')
        writer.writerow(headers)

        for item in tweets:
            analysis = TextBlob(item.text)
            if analysis.sentiment[0]>0:
                sent_class = 'Positive'
            elif analysis.sentiment[0]<0:
                sent_class = 'Negative'
            else:
                sent_class = 'Neutral'

            writer.writerow([item.id,
                             item.text,
                             item.created_at,
                             item.lang,
                             sent_class])
    csvfile.close()
```

```python
def processing(screen_name):
    with open(screen_name+".csv", "rt", encoding="utf-8") as infile, open("m5_rtmax.csv", "wt", encoding="utf-8") as outfile:
        reader = csv.reader(infile)
        writer = csv.writer(outfile)
        conversion = set('_"/.:;,*&^!()+-#$@')
        for row in reader:
            newrow = [''.join(' '  if c in conversion and ord(c)<128 else c for c in entry) for entry in row]
            writer.writerow(newrow)
    csvfile.close()
```
